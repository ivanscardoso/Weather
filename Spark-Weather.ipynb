```python
#Importing all necessary Library
import urllib.request
import json 
import pandas as pd
from pandas.io.json import json_normalize
from sqlalchemy import create_engine

width = [8,5,8,8,8,7,5]
file_name1 = '~/armaghdata.txt'

#Reading file from folder
hist_data1 = pd.read_fwf(file_name1, widths=width, skiprows=5)

#Data Cleansing
hist_data1 = hist_data1[['yyyy', 'mm', 'tmax','rain']]
hist_data1 = hist_data1[hist_data1.tmax != '---']
hist_data1 = hist_data1[hist_data1.rain != '---']
hist_data1 = hist_data1.iloc[1:]
hist_data1.columns = ['year', 'month', 'temp_max', 'rain']
hist_data1[['year', 'month']] = hist_data1[['year', 'month']].astype(int)

#Print dataset content
hist_data1.head(3)
```

```python
#Importing JSON data from API
API_URL1 = "http://api.openweathermap.org/data/2.5/forecast?q=London,UK&units=metric&appid=08c38b4e319d8deaa25bcb4f1f6e0e83"
with urllib.request.urlopen(API_URL1) as url1:
    raw_forecast_data = json.loads(url1.read().decode())
    
    #Here it would Print data in Pretty Json Format
    #print((json.dumps(raw_forecast_data, indent=3)))

#Data Cleansing
forecast_data = json_normalize(raw_forecast_data['list'])
forecast_data = forecast_data[['dt_txt','main.temp_max','rain.3h']]
forecast_data.columns = ['date', 'temp_max', 'rain']

#Print dataset content
forecast_data.head(3)
```

```python
API_URL2 = "http://api.openweathermap.org/data/2.5/weather?q=London,UK&units=metric&appid=08c38b4e319d8deaa25bcb4f1f6e0e83"
with urllib.request.urlopen(API_URL2) as url2:
    raw_current_data = json.loads(url2.read().decode())
current_data = json_normalize(raw_current_data)
current_data = current_data[['dt','main.temp_max']]
current_data.columns = ['date', 'temp_max']
current_data['date'] = pd.to_datetime(current_data['date'],unit='s')
current_data
```

```python
#Create Database Connection
engine = create_engine("mysql://root:password@mysql-spark/MyDB")
db = engine.connect()

#Insert Datasets to Database Tables
hist_data1.to_sql(name='hist_data',con=db,if_exists='append')
hist_data2.to_sql(name='hist_data',con=db,if_exists='append')
hist_data3.to_sql(name='hist_data',con=db,if_exists='append')
hist_data4.to_sql(name='hist_data',con=db,if_exists='append')
hist_data5.to_sql(name='hist_data',con=db,if_exists='append')
hist_data6.to_sql(name='hist_data',con=db,if_exists='append')

forecast_data.to_sql(name='forecast_data',con=db,if_exists='replace')

current_data.to_sql(name='current_data',con=db,if_exists='replace')

#Close Database Connection
db.close()
```
